```python
'''
본질적인 기능만 남긴 최소 형태의 에이전트 코드입니다.
이전 대화의 복잡한 상태 유지(Context)와 반복 루프를 모두 제거하고, 단일 도구와 1회 파싱-실행 구조로 축소했습니다.
'''

# 1. 도구 (Tool): 결정론적 연산
def add(a: int, b: int) -> int: 
    return a + b

# 2. 제어기 (Controller): 파싱 및 실행
def minimal_agent(user_query: str) -> str:
    
    # 3. 판단 (LLM 모사)
    # 실제로는 LLM이 user_query를 읽고 분석하여 아래와 같은 규칙적인 문자열을 생성합니다.
    llm_output = "add,10,20" 
    
    # 제어기의 파싱(Parsing): 문자열을 분리하여 데이터로 변환
    parts = llm_output.split(',')
    command = parts[0] # "add"
    
    # 도구 실행 (Execution)
    if command == "add":
        val1 = int(parts[1])
        val2 = int(parts[2])
        actual_result = add(val1, val2) # 파이썬 함수 호출
        
        return f"최종 결과: {actual_result}"
        
    return "실행할 수 없는 명령입니다."

# 테스트
print(minimal_agent("10과 20을 더해줘"))
```
