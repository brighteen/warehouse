단위: %, 명

산출식
1. 학업 중단률(%) = 학업중단자 / 재적학생수 X 100
2. 학업중단자는 제적학생을 의미하며 미등록, 미복학, 자퇴, 학사경고, 유급, 수업연한초과, 기타사유 등에 의해 학적에서 제외된 학생을 말함
3. 재적학생수는 재학생, 휴학생, 학사학위취득유예생이 포함됨
4. 세종은 2013년부터 학업중단률이 계산됨
5. 조사 기준일
- 학업중단자: 전년도 3월 1일 ~ 당해연도 2월 말
- 재적학생수: 전년도 4월 1일

```python
import pandas as pd

file_path = '2024_연도별 학업중단률.xlsx'
df = pd.read_excel(file_path)
df
```

```python
# 파일 다시 불러오기
df = pd.read_excel(file_path, sheet_name='Sheet0', header=None)

# 1, 2번째 행 가져오기
first_row = df.iloc[0]
second_row = df.iloc[1]

# 새 열 이름 만들기
new_columns = []

for i in range(len(first_row)):
    if i < 2:
        # 1,2번째 열은 첫번째 행 그대로 사용
        new_columns.append(first_row[i])
    else:
        # 나머지 열은 '첫 번째 행' + '_' + '두 번째 행' 조합
        new_columns.append(f"{first_row[i]}_{second_row[i]}")

# 새 헤더 적용하고, 원본 데이터는 2번째 행 이후부터 가져오기
df = df.iloc[2:].reset_index(drop=True)
df.columns = new_columns

df
```

학년도, 시도를 제외한 모든 열(object) → 숫자형(float)으로 변환

중복되던 컬럼명(예: 학업 중단률(%))은 자동으로 _1, _2 붙여서 고유하게 변경

'-' 값은 전부 NaN (결측치)로 변환

필요 없는 'Unnamed' 컬럼 제거

```python
# '-'를 NaN으로 바꾸고, 숫자로 변환할 수 있는 열은 전부 숫자로 변환하는 전처리

# 1. '-'를 NaN으로 치환
df = df.replace('-', pd.NA)

# 2. 학년도, 시도는 그대로 두고 나머지 열만 숫자로 변환
for col in df.columns[2:]:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
# 90% 이상 NaN인 행 제거
total_columns = len(df.columns) - 2
threshold = total_columns * 0.9
nan_counts = df.iloc[:, 2:].isna().sum(axis=1)
df_filtered = df[nan_counts < threshold].reset_index(drop=True)
```

```python
df_filtered.tail(10) # Nan이 90% 이상인 행 제거 후 데이터 확인
```

연도별 학업중단률 확인

```python
# 2018년 이후 데이터만 필터링
df_recent = df_filtered[df_filtered['학년도'] >= 2018].reset_index(drop=True)

df_recent
```

```python
import matplotlib.pyplot as plt

# '학년도'를 숫자로 변환
df_recent['학년도'] = pd.to_numeric(df_recent['학년도'], errors='coerce')

# '학업 중단률(%)_고등교육기관' 기준으로 연도별 평균 학업 중단률을 계산
df_yearly_avg = df_recent.groupby('학년도')[
    [col for col in df_recent.columns if '학업 중단률(%)' in col]
].mean()

# 연도별 전체 평균 학업 중단률 구하기
df_yearly_avg['전체 평균'] = df_yearly_avg.mean(axis=1)

# 그래프 그리기
plt.figure(figsize=(10,6))
plt.plot(df_yearly_avg.index, df_yearly_avg['전체 평균'], marker='o')
# 각 포인트에 주석 추가
for x, y in zip(df_yearly_avg.index, df_yearly_avg['전체 평균']):
    plt.annotate(f"{y:.2f}%", (x, y), textcoords="offset points", xytext=(0, 5), ha='center', fontsize=8)
plt.title('연도별 평균 학업 중단률 추이')
plt.xlabel('학년도')
plt.ylabel('평균 학업 중단률 (%)')
plt.xticks(df_yearly_avg.index)
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
# '시도'별로 최근 데이터(2018년 이후)의 학업 중단률 평균을 계산해보자
# '학업 중단률(%)_고등교육기관'만 일단 기준으로

df_region_avg = df_recent.groupby('시도')[
    [col for col in df_recent.columns if '학업 중단률(%)' in col]
].mean()

# 전체 평균 학업 중단률 컬럼 추가
df_region_avg['전체 평균'] = df_region_avg.mean(axis=1)

# 시도별 학업 중단률 전체 평균으로 정렬
df_region_avg_sorted = df_region_avg.sort_values('전체 평균', ascending=False)

# 그래프 그리기

plt.figure(figsize=(10,6))
plt.bar(df_region_avg_sorted.index, df_region_avg_sorted['전체 평균'])
for x, y in zip(df_region_avg_sorted.index, df_region_avg_sorted['전체 평균']):
    plt.annotate(f"{y:.2f}%", (x, y), textcoords="offset points", xytext=(0, 5), ha='center', fontsize=8)
plt.xticks(rotation=45, ha='right')
plt.title('시도별 평균 학업 중단률 (2018년 이후)')
plt.xlabel('시도')
plt.ylabel('평균 학업 중단률 (%)')
plt.grid(axis='y')
plt.tight_layout()
plt.show()

```
