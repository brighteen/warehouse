```python
import matplotlib.pyplot as plt

# 분석 및 시각화 함수
def plot_box_by_category(x_col, y_col='자퇴학생비율'):
    plt.figure(figsize=(10, 6))
    df.boxplot(column=y_col, by=x_col)
    plt.title(f'{x_col}별 {y_col} 분포')
    plt.suptitle('')
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.show()
```

```python
# 1. 기준연도별 자퇴학생비율
plot_box_by_category('기준연도')
```

```python
# 2. 학교종류별 자퇴학생비율
plot_box_by_category('학교종류')
```

```python
# 3. 설립구분별 자퇴학생비율
plot_box_by_category('설립구분')
```

```python
# 4. 지역별 자퇴학생비율
plot_box_by_category('지역')
```

```python
import seaborn as sns
import numpy as np

# 분석용 숫자형 컬럼만 추출
numeric_cols = ['재적학생', '자퇴학생', '자퇴학생비율', '신입생_자퇴학생', '신입생_자퇴학생비율']
corr_matrix = df[numeric_cols].corr()

# 상관계수 히트맵 그리기
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('자퇴율 관련 변수 상관관계 분석')
plt.show()

# 자퇴학생비율(y)과 각 변수 간 상관계수만 따로 보기
correlations_with_target = corr_matrix['자퇴학생비율'].drop('자퇴학생비율').sort_values(ascending=False)
correlations_with_target
```
