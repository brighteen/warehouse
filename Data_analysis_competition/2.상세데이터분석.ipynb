# 2. 지역별 중도탈락률, 자퇴율 차이 분석

```python
# 지역별 집계
df_region = df.groupby('지역').agg({
    '재적학생수_A': 'sum',
    '중도탈락_계': 'sum',
    '중도탈락_자퇴': 'sum',
    '재적학생수_신입생_Ap': 'sum',
    '중도탈락_계_신입생': 'sum',
    '중도탈락_자퇴_신입생': 'sum'
}).reset_index()

# 지역별 비율 계산
df_region['기존생_중도탈락률(%)'] = (df_region['중도탈락_계'] / df_region['재적학생수_A']) * 100
df_region['기존생_자퇴율(%)'] = (df_region['중도탈락_자퇴'] / df_region['중도탈락_계']) * 100
df_region['신입생_중도탈락률(%)'] = (df_region['중도탈락_계_신입생'] / df_region['재적학생수_신입생_Ap']) * 100
df_region['신입생_자퇴율(%)'] = (df_region['중도탈락_자퇴_신입생'] / df_region['중도탈락_계_신입생']) * 100
```

### 상관관계 분석 결과 해석:

1. 중도탈락비율과 각 사유별 연관성 순위:
   - 위 상관계수를 통해 어떤 사유가 중도탈락비율과 가장 강한 상관관계를 가지는지 확인할 수 있습니다.
   - 양의 상관관계는 해당 사유의 비율이 높을수록 전체 중도탈락비율도 높아진다는 것을 의미합니다.
   - 음의 상관관계는 해당 사유의 비율이 높을수록 전체 중도탈락비율이 낮아진다는 것을 의미합니다.

2. 사유별 평균 비율:
   - 각 사유가 전체 중도탈락에서 차지하는 평균적인 비중을 보여줍니다.
   - 이를 통해 가장 빈번한 중도탈락 사유를 파악할 수 있습니다.

3. 시사점:
   - 상관계수가 높은 사유들은 중도탈락 예방을 위해 우선적으로 관리해야 할 요인입니다.
   - 평균 비율이 높은 사유들은 현재 가장 많이 발생하는 문제로, 즉각적인 대응이 필요한 영역입니다.
