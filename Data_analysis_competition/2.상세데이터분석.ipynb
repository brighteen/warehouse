```python
# 1. 컬럼 이름 정리
# 첫 번째 행에서 '사유별 중도탈락 학생' 이후 열들은 사유별 탈락자 수
new_columns = [
    '기준연도', '학교종류', '설립구분', '지역', '상태', '학교', '재적학생수_A', 
    '중도탈락_계', '중도탈락_미등록', '중도탈락_미복학', '중도탈락_자퇴', '중도탈락_학사경고',
    '중도탈락_학생활동', '중도탈락_유급제적', '중도탈락_수업연한초과', '중도탈락_기타',
    '중도탈락비율(%)_기존생', '재적학생수_신입생_Ap', 
    '중도탈락_계_신입생', '중도탈락_미등록_신입생', '중도탈락_미복학_신입생', '중도탈락_자퇴_신입생',
    '중도탈락_학사경고_신입생', '중도탈락_학생활동_신입생', '중도탈락_유급제적_신입생',
    '중도탈락_수업연한초과_신입생', '중도탈락_기타_신입생', '중도탈락비율(%)_신입생'
]

len(new_columns) # 28
```

```python
# 열 수가 맞으면 교체
df.columns = new_columns

# 2. 의미 없는 열 삭제 (이미 없는 상태로 가정)

# 3. 수치 데이터 숫자형 변환
# '학교명' 같은 것은 제외하고, 수치형 컬럼만 변환
numeric_cols = [col for col in df.columns if '중도탈락' in col or '재적학생수' in col]
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.head()
```

```python
# 모든 컬럼이 NaN인 행 삭제
df = df.dropna(how='all').reset_index(drop=True)
df.head()
```

```python
df.shape
```

```python
# 기존생 자퇴율 계산
df['기존생_자퇴율(%)'] = (df['중도탈락_자퇴'] / df['중도탈락_계']) * 100 # 새 컬럼 추가

# 전체 평균 자퇴율 계산
avg_existing_dropout = df['기존생_자퇴율(%)'].mean(skipna=True) # skipna=True는 NaN을 무시하고 평균 계산

avg_existing_dropout
```

```python
# 신입생 자퇴율 계산
df['신입생_자퇴율(%)'] = (df['중도탈락_자퇴_신입생'] / df['중도탈락_계_신입생']) * 100

avg_freshman_dropout = df['신입생_자퇴율(%)'].mean(skipna=True)

avg_freshman_dropout
```

### 1. 기존 재적학생 중도탈락의 주된 사유는 '자퇴'
- 전체 중도탈락 학생 10명 중 약 6명은 자퇴로 학업을 중단
- 즉, "학업 중단" = "자퇴" 라고 볼 수 있는 강한 근거가 됨
- 다만 나머지 40% 정도는 미등록, 미복학, 학사경고 등 다양한 다른 사유도 함께 존재

### 2. 신입생은 훨씬 더 심각
- 신입생의 경우, 중도탈락자 10명 중 8명 이상(82%)이 자퇴
- 신입생 초기 적응 실패가 학업 중단의 핵심 요인이라는 의미.

학업 중단률 관리 = 자퇴율 관리 / 특히 신입생 대상 관리가 매우 중요

신입생 적응 프로그램 강화 필요 / 오리엔테이션, 학습/심리 상담, 튜터링 제도 강화

신입생 자퇴율 높은 대학별 집중 관리 필요 / 대학별/계열별 분석 통해 집중 지원 대상 식별

# 1. 대학별 재적학생 자퇴율, 신입생 자퇴율 정렬

```python
df_univ_dropout = df[['학교', '지역', '재적학생수_A', '중도탈락_계', '중도탈락_자퇴', 
                                     '재적학생수_신입생_Ap', '중도탈락_계_신입생', '중도탈락_자퇴_신입생']].copy()
# 자퇴율 계산
df_univ_dropout['기존생_자퇴율(%)'] = (df_univ_dropout['중도탈락_자퇴'] / df_univ_dropout['중도탈락_계']) * 100
df_univ_dropout['신입생_자퇴율(%)'] = (df_univ_dropout['중도탈락_자퇴_신입생'] / df_univ_dropout['중도탈락_계_신입생']) * 100

# 대학별 정렬
df_univ_sorted = df_univ_dropout.sort_values(by='기존생_자퇴율(%)', ascending=False)
```

# 2. 지역별 중도탈락률, 자퇴율 차이 분석

```python
# 지역별 집계
df_region = df.groupby('지역').agg({
    '재적학생수_A': 'sum',
    '중도탈락_계': 'sum',
    '중도탈락_자퇴': 'sum',
    '재적학생수_신입생_Ap': 'sum',
    '중도탈락_계_신입생': 'sum',
    '중도탈락_자퇴_신입생': 'sum'
}).reset_index()

# 지역별 비율 계산
df_region['기존생_중도탈락률(%)'] = (df_region['중도탈락_계'] / df_region['재적학생수_A']) * 100
df_region['기존생_자퇴율(%)'] = (df_region['중도탈락_자퇴'] / df_region['중도탈락_계']) * 100
df_region['신입생_중도탈락률(%)'] = (df_region['중도탈락_계_신입생'] / df_region['재적학생수_신입생_Ap']) * 100
df_region['신입생_자퇴율(%)'] = (df_region['중도탈락_자퇴_신입생'] / df_region['중도탈락_계_신입생']) * 100
```

```python
df_region
```

```python
# 지역별 정렬
df_region_sorted = df_region.sort_values(by='기존생_중도탈락률(%)', ascending=False)
df_region_sorted
```
