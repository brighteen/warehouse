```python
import pandas as pd

file_path = rf'C:\Users\Admin\Desktop\25공공교육데이터분석\데이터분석\데이터\중도탈락 학생 현황_학과별\자연과학계열\전처리 전\{file_name}.xlsx'

# 1. 엑셀 파일 읽기 (4·5행을 헤더로)
df = pd.read_excel(file_path, header=[3, 4])

# 2. 멀티인덱스 컬럼을 단일 레벨로 플래튼(flatten)
new_cols = []
for top, sub in df.columns:
    top = top.strip().replace('\n', ' ')
    sub = str(sub).strip()
    # sub 레벨이 'Unnamed'이면 상위 레벨 이름만 사용
    if sub.startswith('Unnamed'):
        new_cols.append(top)
    else:
        new_cols.append(f"{top}_{sub}")
df.columns = new_cols

# 3. 필요한 컬럼만 선택 & 이름 간소화
df_selected = df[[
    '기준연도',
    '학교종류',
    '설립구분',
    '지역',
    '학교',                 # 원본에서 '학교'가 나옵니다
    '학과 (전공)',         # 원본에서 '학과 (전공)' 혹은 '학과 (전공)_<something>' 컬럼
    '재적학생 (A)',
    '사유별 중도탈락 학생_자퇴',
    '중도탈락학생비율(%) (B/A) × 100'
]].copy()

# 4. 컬럼명 최종 리네임
df_selected = df_selected.rename(columns={
    '학교': '학교명',
    '학과 (전공)': '학과',
    '재적학생 (A)': '재적학생',
    '사유별 중도탈락 학생_자퇴': '자퇴학생',
    '중도탈락학생비율(%) (B/A) × 100': '자퇴학생비율'
})
```

```python
import pandas as pd
from pathlib import Path

# 1. 처리할 폴더 경로 지정 (예: 인문사회계열\전처리 전)
input_dir = Path(
    r'C:\Users\Admin\Desktop\25공공교육데이터분석\데이터분석\데이터'
    r'\중도탈락 학생 현황_학과별\공학계열\전처리 전'
)

# 2. 출력 폴더 경로 자동 생성 (전처리 전 → 전처리 후)
output_dir = Path(str(input_dir).replace('전처리 전', '전처리 후'))
output_dir.mkdir(parents=True, exist_ok=True)

# 3. 폴더 내 엑셀 파일 순회
for file_path in input_dir.glob('*.xlsx'):
    # — 읽기 (4·5행을 멀티헤더로)
    df = pd.read_excel(file_path, header=[3, 4])

    # — 컬럼 플래튼(flatten)
    flat_cols = []
    for top, sub in df.columns:
        top = top.strip().replace('\n', ' ')
        sub = str(sub).strip()
        if sub.startswith('Unnamed'):
            flat_cols.append(top)
        else:
            flat_cols.append(f"{top}_{sub}")
    df.columns = flat_cols

    # — 관심 컬럼만 선택
    df_sel = df[[
        '기준연도',
        '학교종류',
        '설립구분',
        '지역',
        '학교',                   # → 나중에 '학교명'으로 리네임
        '학과 (전공)',           # → 나중에 '학과'로 리네임
        '재적학생 (A)',          # → 나중에 '재적학생'으로 리네임
        '사유별 중도탈락 학생_자퇴',  # → '자퇴학생'
        '중도탈락학생비율(%) (B/A) × 100'  # → '자퇴학생비율'
    ]].copy()

    # — 컬럼명 간소화
    df_sel = df_sel.rename(columns={
        '학교': '학교명',
        '학과 (전공)': '학과',
        '재적학생 (A)': '재적학생',
        '사유별 중도탈락 학생_자퇴': '자퇴학생',
        '중도탈락학생비율(%) (B/A) × 100': '자퇴학생비율'
    })

    # — 처리 완료된 파일 저장
    out_path = output_dir / file_path.name
    df_sel.to_excel(out_path, index=False)
    print(f"Processed → {out_path}")
```
