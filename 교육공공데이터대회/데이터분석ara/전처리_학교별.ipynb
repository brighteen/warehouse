```python
file_name = "2022_대학_자연과학계열"
```

```python
import pandas as pd

# 1. 엑셀 파일 읽기 (4번째·5번째 행을 헤더로 사용)
df = pd.read_excel(rf'C:\Users\Admin\Desktop\25공공교육데이터분석\데이터분석\데이터\중도탈락 학생 현황_학교별평균값\전처리 전\{file_name}.xlsx', header=[3, 4])

# 2. 멀티인덱스 컬럼을 단일 레벨로 플래튼(flatten)
new_cols = []
for top, sub in df.columns:
    if pd.isna(sub):
        new_cols.append(str(top).strip())
    else:
        new_cols.append(f"{top.strip()}_{sub.strip()}")
df.columns = new_cols

# 3. 관심 있는 컬럼만 남기고 이름 간소화
df = df.rename(columns={
    '기준연도_Unnamed: 0_level_1': '기준연도',
    '학교종류_Unnamed: 1_level_1': '학교종류',
    '설립구분_Unnamed: 2_level_1': '설립구분',
    '지역_Unnamed: 3_level_1': '지역',
    '학교_Unnamed: 5_level_1': '학교명',
    '재적학생\n(A)_Unnamed: 6_level_1': '재적학생',
    '사유별 중도탈락 학생_자퇴': '자퇴학생',
    '중도탈락학생비율(%)\n(B/A) × 100_Unnamed: 16_level_1': '자퇴학생비율',
    '사유별 중도탈락 학생(신입생)_자퇴': '신입생_자퇴학생',
    '중도탈락학생(신입생)비율(%)\n(B`/A`) × 100_Unnamed: 27_level_1': '신입생_자퇴학생비율'
})

# 4. 필요한 열만 선택
selected_cols = [
    '기준연도', '학교종류', '설립구분', '지역', '학교명',
    '재적학생', '자퇴학생', '자퇴학생비율',
    '신입생_자퇴학생', '신입생_자퇴학생비율'
]
df_selected = df[selected_cols]
```
