```python
import pandas as pd

# 데이터 로드
strong_df = pd.read_excel('datas/STRONG_최종.xlsx')
course_history_df = pd.read_excel('datas/교양교과_수강정보_최종.xlsx')
course_list_df = pd.read_excel('datas/교양교과목록_최종.xlsx')
```

```python


# 1. STRONG 역량 진단 데이터 통계
strong_stats = {
    "총 학생 수": strong_df['학번'].nunique(),
    "단과대학 수": strong_df['COLG_NM'].nunique(),
    "학과 수": strong_df['SUST_NM'].nunique(),
    "역량 점수 요약": strong_df.filter(like='_T_SC').describe().round(2)
}

# 2. 교양교과 수강 이력 데이터 통계
history_stats = {
    "총 수강 건수": len(course_history_df),
    "수강 학생 수": course_history_df['HAKBEON'].nunique(),
    "개설된 과목 수": course_history_df['GWAMOK_NO'].nunique(),
    "평점 평균 요약": course_history_df['PYEONGJEOM_AVG'].describe().round(2),
    "가장 많이 수강한 과목 TOP 5": course_history_df['GWAMOK_KOR_NM'].value_counts().head(5)
}

# 3. 교양교과 목록 데이터 통계
list_stats = {
    "등록된 총 교과목 수": course_list_df['GWAMOK_NO'].nunique(),
    "학점 단위 분포": course_list_df['HAKJEOM'].value_counts().sort_index()
}

# 결과 출력을 위한 데이터프레임 표시 (텍스트 요약으로 제공)
strong_stats, history_stats, list_stats
```
